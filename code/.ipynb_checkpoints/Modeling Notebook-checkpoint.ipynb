```python
df.tail(2)
```

```python
df.drop(columns=['wins','loss'], inplace=True)
```

## K-Means Clustering

```python
df2 = df.copy()
```

```python
X = df.iloc[:,3:33]
ss = StandardScaler()
ss.fit(X)
X_scaled = ss.transform(X)
```

```python
km = KMeans(n_clusters=50, random_state= 1906)
km.fit(X_scaled)
```

```python
km.labels_
```

```python
df2['cluster'] = km.labels_
```

```python
df2[['team','year','win_pct','off_pts','def_pts','cluster','league']].sort_values(by='cluster',ascending=False).head(50)
```
