## Column Alignment

### NFL
---

* The columns between both the NFL & College DataFrames have different orders. So I am going to realign them to be better suited to be joined later on.

* Since we will be comparing the NFL to the College stats, I would like for my stats to be on a per game basis to make sure we are comparing apples to apples as much as possible. (NFL plays more games than College so their total stats would be more than college's).

```python
current_NFL['comp']
```

```python
nfl_per_cols = ['points_scored','comp', 'attempts','pass_tds','int','rush_att','rush_tds','fumbles',
                'points_allow','opp_yards','opp_plays','turnover','opp_fumb','opp_comp','opp_pass_att',
                'opp_pass_yds','opp_pass_td','def_int','opp_rush_att','opp_rush_yds','opp_rush_td']
```

Recalculating each column to a per game basis by dividing by the number of games the NFL plays. I am doing this to properly match the stats with the college DataFrame. ### Plotting Function

```python
# Reference: https://realpython.com/python-matplotlib-guide/
# Helps place a title inside of the plot. Is useful when placing multiple subplots inside of a grid.
def add_titlebox(ax, text):
        ax.text(.55, .9, text,
            horizontalalignment='center',
            transform=ax.transAxes,
            bbox=dict(facecolor='white', alpha=0.6),
            fontsize=12.5)
        return ax
```

```python
def stat_dist(df, list_of_columns, list_of_categories):
    
    for i, column in enumerate(list_of_columns):
        title1 = list_of_categories[i] + ' by League'
        title2 = 'Histogram of NFL ' + list_of_categories[i]
        title3 = 'Histogram of College ' + list_of_categories[i]
        
        gridsize = (4, 2)
        fig = plt.figure(figsize=(16, 8))
        ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=1, rowspan=4)
        ax2 = plt.subplot2grid(gridsize, (0, 1), colspan=1, rowspan=2)
        ax3 = plt.subplot2grid(gridsize, (2, 1), colspan=1, rowspan=2)

        ax1.set_title(title1, fontsize=14)

        sns.boxplot(x=df["league"], y=df[column], ax=ax1)
        rec_df[df['league'] == 'NFL'][column].hist(ax=ax2, bins=25)
        rec_df[df['league'] == 'College'][column].hist(ax=ax3, bins=25);
        
        # use previously created function to add titles to the histograms
        add_titlebox(ax2, title2)
        add_titlebox(ax3, title3)
        
    return print(f"Plotting is complete. There are {len(list_of_columns)} plots")
```

## Scaling Columns

```python
current_NFL.reset_index(drop=True, inplace=True)
```

```python
current_College.reset_index(drop=True, inplace=True)
```

I am resetting the index to both league's DataFrames to properly update the values of each with the scaled values.

```python
current_NFL.head(2)
```

```python
current_College.head(2)
```

The multiple statistical categories have stats of different ranges. When comparing different categories to each other this will provide not so preferable results. In addition to this, when we look at one category and compare college to NFL, there are large descrepancies between the two. 

Since my goal is to provide a relative college football team to a specific NFL team, the stats having different spreads and being on different scales will not always accomplish this goal. My initial recommender has evidence of this issue. Top NFL teams are returning middle of the pack college football teams.
* I would like to see the top NFL teams being compared to the top College Teams overall and in specific statistical categories.

To solve these problems I will apply a Min Max Scale to both the NFL & College Football DataFrames seperately.
* This will first make sure that all of my categories are on a scale/range between 0 and 1.
* Secondly I will have each league top stats closer to 1 and the worse stats closest to 0, on their own scales. So when the pairwise distances are calculated they will be compared accordingly. (top performers of NFL to top performers of College)


### NFL Scaling

```python
no_scale = ['team','year','games_played','wins','loss','win_pct','comp_pct','opp_comp_pct','league']
```

```python
# Selecting only the columns I need to apply Min Max Scale to.
scale_cols = current_NFL.loc[:,~current_NFL.columns.isin(no_scale)]
```

```python
mm = MinMaxScaler()
scaled = mm.fit_transform(scale_cols)
```

```python
scaled_df = pd.DataFrame(scaled, columns=scale_cols.columns)
scaled_df.head()
```

```python
current_NFL.update(scaled_df)
```

```python
current_NFL.head(10)
```

### College Scaling

```python
# Selecting only the columns I need to apply Min Max Scale to.
scale_colsC = current_College.loc[:,~current_College.columns.isin(no_scale)]
```

```python
scaledC = mm.fit_transform(scale_colsC)
```

```python
scaled_df2 = pd.DataFrame(scaledC, columns=scale_colsC.columns)
scaled_df2.head()
```

```python
current_College.update(scaled_df2)
```

```python
current_College.head(10)
```

```python
# Final DataFrame for recommender system - with scaled data.
scaled_rec_df = pd.concat([current_NFL, current_College],ignore_index=True)
scaled_rec_df.head(2)
```
