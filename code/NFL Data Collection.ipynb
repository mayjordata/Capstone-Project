# Selenium Solution

```python
# Link to the main page
url = 'https://www.pro-football-reference.com/years/'
```

```python
driver = webdriver.Chrome(executable_path="./chromedriver/macos/chromedriver")
```

```python
driver.get(url)
```

```python
# creating soup for the inital page documented in the "url" variable.
soup = BeautifulSoup(driver.page_source)
```

### Finding Years

```python
# Finding the table on the page that contains all of the years in the NFL.
years_table = soup.find('table', {'id': 'years'})
```

```python
# Obtaining each numerical year
years_list = []

for row in years_table.find_all('tr')[1:52]:
    if row.find('th').text == 'Year':
        pass
    else:
        year = row.find('th').text
        years_list.append(year)
```

This loop collects all of the years in the NFL I would like to collect data from. Dating back to 1970.

### Data Scrapper per Year

```python
# Looping through each year page

# Empty containeer for each team dictionary for overall stats
teams = []


# Offensive stats DataFrames
# Initiating DataFrame for passing stats with column names
pass_df = pd.DataFrame(columns=
                  ['team','year','game','comp','attempts','comp_perc','pass_yards','pass_tds','pass_td_perc',
                   'int','int_perc','pass_long','pass_y/a','y/c','pass_y/g','QB_rate','sacks','sack_yds',
                   'sack_perc','qb_comebacks','qb_gwd','pass_exp'])

# Initiating DataFrame for rushing stats with column names
rush_df = pd.DataFrame(columns=
                  ['team','year','game','rush_att','rush_yards','rush_tds','rush_long',
                   'rush_y/a','rush_y/g','fumbles','rush_exp'])

# Initiating DataFrame for defensive stats with column names
defense_df = pd.DataFrame(columns=
                  ['team','year','game','points_allow','opp_yards','opp_plays','opp_y/p','turnover','opp_fumb','opp_1stD',
                   'opp_comp','opp_pass_att','opp_pass_yds','opp_pass_td','def_int','opp_Ny/a','opp_pass_1stD','opp_rush_att','opp_rush_yds',
                   'opp_rush_td','opp_rush_y/a','opp_rush_1stD','def_pen','def_pen_yds','def_pen_1stD','opp_score_pct','turnover_pct','def_exp'])


start = datetime.now()

for year in years_list: # Looping through each year from current year (2019) to 1970.
    
    print(f"scrapping year: {year}")
    
    # Link to specific year for selenium to click on. Using the original url combined with the year to access the specific year.
    driver.get(url + year)
    
    # creating soup for current year in loop
    year_soup = BeautifulSoup(driver.page_source)
    
    # Seek AFC Table and collect Data
    AFC_table = year_soup.find('div', {'id': 'div_AFC'})
    sleep(3)
    
    # looping throuh each row in the current year AFC table and collecting data
    for row in AFC_table.find_all('tr')[1:]:
        team = {}
        tm = row.find('th')
        wins = row.find('td', {'data-stat': 'wins'})
        loss = row.find('td', {'data-stat': 'losses'})
        tie = row.find('td', {'data-stat': 'ties'})
        wlp = row.find('td', {'data-stat': 'win_loss_perc'})
        pts = row.find('td', {'data-stat': 'points'})
        ptsA = row.find('td', {'data-stat': 'points_opp'})
        dpts = row.find('td', {'data-stat': 'points_diff'})
        moV = row.find('td', {'data-stat': 'mov'})
        sos = row.find('td', {'data-stat': 'sos_total'})
        srs = row.find('td', {'data-stat': 'srs_total'})
        osrs = row.find('td', {'data-stat': 'srs_offense'})
        dsrs = row.find('td', {'data-stat': 'srs_defense'})


        if tm:
            team['team'] = tm.text.replace('*','').replace('+','')
            team['year'] = year
        if wins:
            team['wins'] = wins.text
        if loss:
            team['losses'] = loss.text
        if tie:
            team['ties'] = tie.text
        if wlp:
            team['win_loss_perc'] = wlp.text
        if pts:
            team['points_scored'] = pts.text
        if ptsA:
            team['opp_points'] = ptsA.text
        if dpts:
            team['points_diff'] = dpts.text
        if moV:
            team['margin_of_victory'] = moV.text
        if sos:
            team['SoS'] = sos.text 
        if srs:
            team['SRS'] = srs.text
        if osrs:
            team['OSRS'] = osrs.text
        if dsrs:
            team['DSRS'] = dsrs.text


        else:
            continue

        # appending created dictionary of current year to teams list
        teams.append(team)
    
    # Seek NFC Table and collect Data
    NFC_table = year_soup.find('div', {'id': 'div_NFC'})
    sleep(3)
    
    # looping throuh each row in the current year NFC table and collecting data
    for row in NFC_table.find_all('tr')[1:]:
        team = {}
        tm = row.find('th')
        wins = row.find('td', {'data-stat': 'wins'})
        loss = row.find('td', {'data-stat': 'losses'})
        tie = row.find('td', {'data-stat': 'ties'})
        wlp = row.find('td', {'data-stat': 'win_loss_perc'})
        pts = row.find('td', {'data-stat': 'points'})
        ptsA = row.find('td', {'data-stat': 'points_opp'})
        dpts = row.find('td', {'data-stat': 'points_diff'})
        moV = row.find('td', {'data-stat': 'mov'})
        sos = row.find('td', {'data-stat': 'sos_total'})
        srs = row.find('td', {'data-stat': 'srs_total'})
        osrs = row.find('td', {'data-stat': 'srs_offense'})
        dsrs = row.find('td', {'data-stat': 'srs_defense'})

        if tm:
            team['team'] = tm.text.replace('*','').replace('+','')
            team['year'] = year
        if wins:
            team['wins'] = wins.text
        if loss:
            team['losses'] = loss.text
        if tie:
            team['ties'] = tie.text
        if wlp:
            team['win_loss_perc'] = wlp.text
        if pts:
            team['points_scored'] = pts.text
        if ptsA:
            team['opp_points'] = ptsA.text
        if dpts:
            team['points_diff'] = dpts.text
        if moV:
            team['margin_of_victory'] = moV.text
        if sos:
            team['SoS'] = sos.text 
        if srs:
            team['SRS'] = srs.text
        if osrs:
            team['OSRS'] = osrs.text
        if dsrs:
            team['DSRS'] = dsrs.text


        else:
            continue

        # appending created dictionary of current year to teams list
        teams.append(team)
        
    # Seek the passing table in the current NFL year.
    pass_table = year_soup.find('table', {'id': 'passing'})
    
    # Looping through each row of the current year passing table and collecting data
    for row in pass_table.find('tbody').find_all('tr'):
        team = row.find('td').text
        year = year
        game = row.find('td', {'data-stat': 'g'}).text
        comp = row.find('td', {'data-stat': 'pass_cmp'}).text
        attempts = row.find('td', {'data-stat': 'pass_att'}).text
        comp_perc = row.find('td', {'data-stat': 'pass_cmp_perc'}).text
        pass_yards = row.find('td', {'data-stat': 'pass_yds'}).text
        pass_tds = row.find('td', {'data-stat': 'pass_td'}).text
        pass_td_perc = row.find('td', {'data-stat': 'pass_td_perc'}).text
        intercept = row.find('td', {'data-stat': 'pass_int'}).text
        int_perc = row.find('td', {'data-stat': 'pass_int_perc'}).text
        pass_long = row.find('td', {'data-stat': 'pass_long'}).text
        pass_ya = row.find('td', {'data-stat': 'pass_yds_per_att'}).text
        yc = row.find('td', {'data-stat': 'pass_yds_per_cmp'}).text
        pass_yg = row.find('td', {'data-stat': 'pass_yds_per_g'}).text
        qb_rate = row.find('td', {'data-stat': 'pass_rating'}).text
        sacks = row.find('td', {'data-stat': 'pass_sacked'}).text
        sack_yds = row.find('td', {'data-stat': 'pass_sacked_yds'}).text
        sack_perc = row.find('td', {'data-stat': 'pass_sacked_perc'}).text
        qb_cmbk = row.find('td', {'data-stat': 'comebacks'}).text
        qb_gwd = row.find('td', {'data-stat': 'gwd'}).text
        try:
            pass_exp = row.find('td', {'data-stat': 'exp_pts_pass'}).text
        except:
            pass_exp = None
        
        
        # adding each row of stats to the passing DataFrame
        pass_df.loc[len(pass_df)] = [team, year, game, comp, attempts, comp_perc,
                      pass_yards, pass_tds, pass_td_perc, intercept,
                      int_perc, pass_long, pass_ya, yc, pass_yg,
                      qb_rate, sacks, sack_yds, sack_perc, qb_cmbk, qb_gwd, pass_exp]
    
    # Seek the rushing table in the current NFL year.
    rush_table = year_soup.find('table', {'id': 'rushing'})
    
    
    # Looping through each row of the current year rushing table and collecting data
    for row in rush_table.find('tbody').find_all('tr'):
        team = row.find('td').text
        year = year
        game = row.find('td', {'data-stat': 'g'}).text
        attempts = row.find('td', {'data-stat': 'rush_att'}).text
        rush_yards = row.find('td', {'data-stat': 'rush_yds'}).text
        rush_tds = row.find('td', {'data-stat': 'rush_td'}).text
        rush_long = row.find('td', {'data-stat': 'rush_long'}).text
        rush_ya = row.find('td', {'data-stat': 'rush_yds_per_att'}).text
        rush_yg = row.find('td', {'data-stat': 'rush_yds_per_g'}).text
        fumb = row.find('td', {'data-stat': 'fumbles'}).text
        try:
            rush_exp = row.find('td', {'data-stat': 'exp_pts_rush'}).text
        except:
            rush_exp = None
        
        
        # adding each row of stats to the rushing DataFrame
        rush_df.loc[len(rush_df)] = [team, year, game, attempts, rush_yards, rush_tds, 
                                     rush_long, rush_ya, rush_yg, fumb, rush_exp]
    
    sleep(5)
    
    
    
    # Navigate to the Defensive page
    # using the original url combined with the year to access the specific year, along with "/opp.htm"
    driver.get(url + year + '/opp.htm')
    
    # creating soup for current year's defensive page in the loop
    year_soup = BeautifulSoup(driver.page_source)
    
    # Seek the defense table in the current NFL year.
    def_table = year_soup.find('table', {'id': 'team_stats'})
    
    # Looping through each row of the current year defense table and collecting data
    for row in def_table.find('tbody').find_all('tr'):
        team = row.find('td').text
        year = year
        game = row.find('td', {'data-stat': 'g'}).text
        points = row.find('td', {'data-stat': 'points'}).text
        yards = row.find('td', {'data-stat': 'total_yards'}).text
        plays = row.find('td', {'data-stat': 'plays_offense'}).text
        yds_per = row.find('td', {'data-stat': 'yds_per_play_offense'}).text
        to = row.find('td', {'data-stat': 'turnovers'}).text
        fumb = row.find('td', {'data-stat': 'fumbles_lost'}).text
        first_d = row.find('td', {'data-stat': 'first_down'}).text
        comp = row.find('td', {'data-stat': 'pass_cmp'}).text
        pass_att = row.find('td', {'data-stat': 'pass_att'}).text
        pass_yds = row.find('td', {'data-stat': 'pass_yds'}).text
        pass_tds = row.find('td', {'data-stat': 'pass_td'}).text
        pass_int = row.find('td', {'data-stat': 'pass_int'}).text
        pass_nya = row.find('td', {'data-stat': 'pass_net_yds_per_att'}).text
        pass_fd = row.find('td', {'data-stat': 'pass_fd'}).text
        rush_att = row.find('td', {'data-stat': 'rush_att'}).text
        rush_yds = row.find('td', {'data-stat': 'rush_yds'}).text
        rush_tds = row.find('td', {'data-stat': 'rush_td'}).text
        rush_yd_att = row.find('td', {'data-stat': 'rush_yds_per_att'}).text
        rush_fd = row.find('td', {'data-stat': 'rush_fd'}).text
        penalty = row.find('td', {'data-stat': 'penalties'}).text
        penalty_yds = row.find('td', {'data-stat': 'penalties_yds'}).text
        pen_fd = row.find('td', {'data-stat': 'pen_fd'}).text
        score_pct = row.find('td', {'data-stat': 'score_pct'}).text
        turnover_pct = row.find('td', {'data-stat': 'turnover_pct'}).text
        try:
            def_exp = row.find('td', {'data-stat': 'exp_pts_def_tot'}).text
        except:
            def_exp = None
        
        
        # adding each row of stats to the defensive DataFrame
        defense_df.loc[len(defense_df)] = [team, year, game, points, yards, plays, yds_per, to, fumb, first_d, comp,
                                          pass_att, pass_yds, pass_tds, pass_int, pass_nya, pass_fd, rush_att, rush_yds, rush_tds,
                                          rush_yd_att, rush_fd, penalty, penalty_yds, pen_fd, score_pct, turnover_pct, def_exp]
    
    sleep(10)
     
team_df = pd.DataFrame(teams)
stop = datetime.now()
print(f'The total time passed during scrape (hh:mm:ss:ms) is: {stop - start}')
```
